```python
import pandas as pd
```

```python
afg= pd.read_excel("Exercise_data_pandas.xlsx","Afghanistan")
alb= pd.read_excel("Exercise_data_pandas.xlsx","Albania")
alg= pd.read_excel("Exercise_data_pandas.xlsx","Algeria")
```

```python
afg
alb
alg
```

```python
afg.index = pd.to_datetime(afg.date)
alb.index = pd.to_datetime(alb.date)
alg.index = pd.to_datetime(alg.date)
```

```python
# afg
# alg
alb
```

```python
world_cases= pd.DataFrame(afg.Afghanistan)
world_cases["Albania"] = alb.Albania
world_cases["Algeria"] = alg.Algeria
```

```python
world_cases
```

```python
#Create Multiple line chart
excel_file = "Pandas_WarmUpExercise.xlsx"
writer = pd.ExcelWriter(excel_file, engine="xlsxwriter")
sheet_name = "world_cases"
world_cases.to_excel(writer,sheet_name = sheet_name)


# Get the workbook
workbook = writer.book 

#chartsheet 
chartsheet = workbook.add_chartsheet('world_cases_chart')

#get the sheet name
worksheet = writer.sheets[sheet_name]

# Create a chart object.
chart = workbook.add_chart({'type':'line'})

# configure the series of the chart from the dataframe data
for i in range(1, len(world_cases.columns.values)+1):
    chart.add_series({
        'name'      : [sheet_name, 0, i],
        'values'    : [sheet_name, 1, i, 248, i],
        'categories': [sheet_name, 1, 0, 248, 0]
    })

## add axis labels
chart.set_x_axis({
        'name':'Date',
        'num_font':{'rotation':45}
})
chart.set_y_axis({'name':'Total Cases','major_gridlines':{'visible':False}})

# #insert the chart into the worksheet
chartsheet.set_chart(chart)

writer.save()
```
